### Data Wrangling on Playstore Dataframe





```python
# Write your code to make your dataset analysis ready.

#checking outliers in Ratings columns before considering mean or median value to impute against null values

print(df_playstore["Rating"].describe())

sns.boxplot(x=df_playstore["Rating"])
```

```python
 #we can observe in the above plot that there is a single data which is more than 5 rating ideally it should be 5 or below 5 because rating should 1-5 as per data
df_playstore.drop(df_playstore.loc[df_playstore["Rating"]>5].index,inplace=True) #removing the outlier

#we can observe outliers are present , so we can continue to fill null values with the median
```

```python
#filling the Null values with median of the column
df_playstore["Rating"].fillna(df_playstore["Rating"].median(),inplace=True)

#checking if null value removed or not
print(df_playstore["Rating"].isnull().sum())

#checking shape of the data frame now
print(f"{df_playstore.shape}-shape")
```

```python
#droping remaining rows ("Type","Content Rating",'Current Ver',"Android Ver") containing null values as null values were very less ieven less than 0.5%

df_playstore.dropna(subset=["Type","Content Rating",'Current Ver',"Android Ver"],axis=0,inplace=True)

#checking if null values still exists or not
print(df_playstore.isnull().sum())
```

```python
#converting datatype of Reviews to integer type and Last Update to date time for futher analysis

df_playstore["Reviews"]=df_playstore["Reviews"].astype("int")
df_playstore["Last Updated"]=pd.to_datetime(df_playstore["Last Updated"])
```

```python
#Function to remove m and k from size columns + converting KB to MB

def check(n):
  #if condition checking for m in string and removing same and returning
    if n[-1] not in "Kk":
      string=""
      #for loop used to run through string and check presence of m in element
      for i in n:
        if i not in "mM":
          string+=i
      #returning string after removing m from complete string
      return string
    #eslse condition checking for k and converting to MB and returnig
    else:
      string=""
      #for loop used to run through string and check presence of k in element
      for i in n:
        if i not in "kK":
          string+=i
      string=round(float(string)/1024,1)
      #after conversion to MB retruning string only
      return str(string)



#using apply method with lambda and fucntion to remove m and k elements from all the elements in the column
df_playstore["Size"]=df_playstore.apply(lambda x : check(x["Size"]),axis=1)

#checking the unique value now and we can observe size 'Varies with device' present in columns which needs to be considered before converting size column to numerical
df_playstore["Size"].unique()
```

```python
#replacing all varies with device records to null values
df_playstore["Size"].replace("Varies with device",np.nan,inplace=True)

#converting the entire size column to numerical after replacing with null values
df_playstore["Size"]=df_playstore["Size"].astype("float")

#checking null values in size columns and it's count and percentage to proceed with correct action
```

```python
size_null=df_playstore["Size"].isnull().sum()
size_null_percentage= round(df_playstore["Size"].isnull().sum()/len(df_playstore["Size"]) *100,2)

print(f"Total Null values in size columns is {size_null}  and {size_null_percentage} % ")

#we cannot drope around 14.74% null values , need to impute
```

```python
#checking outliers in Size columns before considering mean or median value to impute against null values

print(df_playstore["Size"].describe())

sns.boxplot(x=df_playstore["Size"])

#we can observe outliers are present , so we can continue to fill null values with the median
```

```python
#filling null values with median of size columns
df_playstore["Size"].replace(np.nan,df_playstore["Size"].median(),inplace=True)

#checking if null values still exists or not
print(df_playstore["Size"].isnull().sum())
```

```python
#fucntion to remove "+" and "," from the elements of installs and converting to integers
def remove(element):
  string=""
  for i in element:
    if i not in "+,":
      string+=i
  return int(string)

#using apply method to with remove fucntion and co
df_playstore["Installs"]=df_playstore["Installs"].apply(remove)
```

```python
#fucntion to remove "$"from price column  and converting to float
def remove_dollar(element):
  string=""
  for i in element:
    if i not in "$":
      string+=i
  return float(string)
''
  #using apply method to with remove fucntion and co
df_playstore["Price"]=df_playstore["Price"].apply(remove_dollar)
```

#### Chart - 3 Category Vs Ratings/Reviews(Bivariate)

```python
# Chart - 3 visualization code
df=df_playstore.copy()
print(f"Maxium Rating {df_playstore.Rating.max()} and Minimum Rating is {df_playstore.Rating.min()}")

#function to categories Rating column to Good(4 and above) Average(3-4) and Poor(1-3)
def Rating(n):
  if n>=1 and n<3:
    return "Poor"
  elif n>=3 and n<4:
    return "Average"
  else:
    return "Good"

df["Cat_Rating"]=df["Rating"].apply(Rating)
df.head(1)
```

```python
#using cross tab to know the count of good ,average and poor ratings respectively
good=pd.crosstab(df["Category"],df["Cat_Rating"]).reset_index().sort_values(by="Good",ascending= False)
Poor=pd.crosstab(df["Category"],df["Cat_Rating"]).reset_index().sort_values(by="Poor",ascending= False)
average=pd.crosstab(df["Category"],df["Cat_Rating"]).reset_index().sort_values(by="Average",ascending= False)
```

```python
fig,axes=plt.subplots(1,3,figsize=(20,6))

#ploting Category vs Good Rating
sns.barplot(y=good["Category"],x=good["Good"],palette="plasma",ax=axes[0])
axes[0].set_title("Category vs Rating(Good)")

#ploting Category vs Poor Rating
sns.barplot(y=Poor["Category"],x=Poor["Poor"],ax=axes[1],palette="RdGy")
axes[1].set_title("Category vs Rating(Poor)")

#ploting Category vs Average Rating
sns.barplot(y=average["Category"],x=average["Average"],ax=axes[2],palette="PuOr")
axes[2].set_title("Category vs Rating(Average)")
plt.tight_layout()
```
